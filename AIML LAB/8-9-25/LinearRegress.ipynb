```python
#WAP to caluculate the pricw of the pizza by using Linear regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

d = np.array([ 6, 8 , 10, 12, 14]).reshape(-1, 1)
p = np.array([ 13, 16, 19, 22, 25])

model = LinearRegression()
model.fit(d, p)
newDiameter = np.array([[21]])
newPrice = model.predict(newDiameter)
print("The price of the new pizza is ", newPrice)
plt.plot(d,p)
plt.title("Linear regression")
plt.xlabel("Diameter----->")
plt.ylabel("Price------->")
plt.show()
```
